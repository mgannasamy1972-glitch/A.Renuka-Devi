```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("ibm-granite/granite-3.3-2b-instruct")
model = AutoModelForCausalLM.from_pretrained("ibm-granite/granite-3.3-2b-instruct")
messages = [
{"role": "user", "content": "Who are you?"},
]
inputs = tokenizer.apply_chat_template(
messages,
add_generation_prompt=True,
tokenize=True,
return_dict=True,
return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs, max_new_tokens=40)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```

```python
from typing import Dict

class ColorPaletteGenerator:
    """Generate visually cohesive color palettes"""

    @staticmethod
    def generate_palette(style: str = "modern") -> Dict[str, str]:

        palettes = {
            "modern": {
                "primary": "#667eea",
                "secondary": "#764ba2",
                "accent": "#f093fb"
            }
        }

        return palettes.get(style.lower(), palettes["modern"])
```

```python
from typing import List, Dict

class HTMLPageBuilder:
    """Build styled HTML pages from descriptions"""

    @staticmethod
    def extract_sections(description: str) -> List[str]:
        """Extract sections from description"""
        sections = []

        section_keywords = {
            'hero': ['hero', 'banner', 'landing'],
            'features': ['features', 'services', 'benefits'],
            'about': ['about', 'team', 'company'],
            'portfolio': ['portfolio', 'gallery', 'work', 'projects'],
            'testimonials': ['testimonial', 'reviews', 'feedback'],
            'pricing': ['pricing', 'plans', 'cost'],
            'contact': ['contact', 'form', 'reach'],
            'footer': ['footer', 'bottom']
        }

        desc_lower = description.lower()

        for section_type, keywords in section_keywords.items():
            for keyword in keywords:
                if keyword in desc_lower:
                    sections.append(section_type)
                    break

        if not sections:
            sections = ['hero', 'features', 'contact', 'footer']

        return sections
```

```python
class CSSGenerator:

    @staticmethod
    def generate_css(palette):
        """Generate comprehensive CSS"""

        css = f"""
        body {{
            background: {palette['background']};
            color: {palette['text_primary']};
            font-family: Arial;
        }}
        """

        return css
```

```python
import gradio as gr
from datetime import datetime

# Simple color palettes
palettes = {
    "modern": {"primary": "#6366f1", "secondary": "#8b5cf6", "accent": "#ec4899", "background": "#0f172a"},
    "ocean": {"primary": "#0284c7", "secondary": "#06b6d4", "accent": "#22d3ee", "background": "#0c4a6e"},
    "forest": {"primary": "#16a34a", "secondary": "#22c55e", "accent": "#4ade80", "background": "#052e16"}
}

history = []

def generate_html(title, description, style):
    if not title or not description:
        return "", "❌ Enter title and description", ""

    palette = palettes.get(style, palettes["modern"])

    html = f"""
<!DOCTYPE html>
<html>
<head>
<title>{title}</title>
<style>
body {{
font-family: Arial;
background:{palette['background']};
color:white;
margin:0;
}}

header {{
background:{palette['primary']};
padding:20px;
text-align:center;
}}

section {{
padding:40px;
}}

button {{
background:{palette['accent']};
border:none;
padding:10px 20px;
color:white;
cursor:pointer;
}}
</style>
</head>

<body>

<header>
<h1>{title}</h1>
</header>

<section>
<p>{description}</p>
<button>Learn More</button>
</section>

</body>
</html>
"""

    history.append({
        "title": title,
        "time": datetime.now().isoformat()
    })

    colors = f"""
Primary: {palette['primary']}
Secondary: {palette['secondary']}
Accent: {palette['accent']}
Background: {palette['background']}
"""

    return html, "✅ HTML Generated", colors


def preview(html):
    return html if html else "<h3>No HTML Generated</h3>"


def palette_info(style):
    p = palettes[style]
    return f"""
Primary: {p['primary']}
Secondary: {p['secondary']}
Accent: {p['accent']}
Background: {p['background']}
"""


with gr.Blocks(title="HTML Quick Styler") as demo:

    gr.Markdown("# 🎨 HTML Quick Styler")

    with gr.Tabs():

        # TAB 1
        with gr.TabItem("Generate"):

            title = gr.Textbox(label="Title")
            desc = gr.Textbox(label="Description", lines=3)

            style = gr.Dropdown(
                ["modern", "ocean", "forest"],
                value="modern",
                label="Style"
            )

            btn = gr.Button("Generate")

            html_output = gr.Code(language="html")
            status = gr.Markdown()
            colors = gr.Markdown()

            btn.click(generate_html,
                      inputs=[title, desc, style],
                      outputs=[html_output, status, colors])

        # TAB 2
        with gr.TabItem("Preview"):

            preview_btn = gr.Button("Show Preview")
            preview_box = gr.HTML()

            preview_btn.click(preview,
                              inputs=[html_output],
                              outputs=[preview_box])

        # TAB 3
        with gr.TabItem("Palettes"):

            style_select = gr.Radio(["modern", "ocean", "forest"])
            palette_btn = gr.Button("Get Colors")
            palette_text = gr.Markdown()

            palette_btn.click(palette_info,
                              inputs=[style_select],
                              outputs=[palette_text])

demo.launch()
```
